```python
data1=data[['GENDER','AGE','SMOKING','CHRONIC DISEASE','ALCOHOL CONSUMING','CHEST PAIN','LUNG_CANCER']]
```

```python
data1
```

```python
from sklearn.preprocessing import LabelEncoder
lebenc=LabelEncoder()
data1['GENDER']=lebenc.fit_transform(data1['GENDER'])
data1
```

```python
from sklearn.preprocessing import LabelEncoder
lebenc=LabelEncoder()
columns_to_encode=['GENDER','LUNG_CANCER']
for column in columns_to_encode:
    data1[column]=lebenc.fit_transform(data1[column])
print(data1)
```

```python
X=data1[['GENDER','AGE','SMOKING','CHRONIC DISEASE','ALCOHOL CONSUMING','CHEST PAIN','LUNG_CANCER']]
X
```

```python
Y=data1[["LUNG_CANCER"]]
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.05,random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.svm import SVC
classifier=SVC(kernel='linear',random_state=0)
classifier.fit(X_train,y_train)
y_pred=classifier.predict(X_test).round()
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
score_lr=round(accuracy_score(y_pred,y_test)*100,2)
print("the accuracy score achieved using DT is:"+str(score_lr)+"%")
```

```python
from sklearn.tree import DecisionTreeClassifier
mdt=DecisionTreeClassifier(criterion="entropy",random_state=0)
mdt=mdt.fit(X_train,y_train)
y_pred1=mdt.predict(X_test).round()
```

```python
y_pred1
```

```python
from sklearn.metrics import accuracy_score
score_lr=round(accuracy_score(y_pred1,y_test)*100,2)
print("the accuracy score achieved using DT is:"+str(score_lr)+"%")
```
